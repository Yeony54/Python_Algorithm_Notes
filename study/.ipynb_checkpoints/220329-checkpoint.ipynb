4줄짜리 표에 왼쪽부터 수를 아래로 1부터 순서대로 적어나간다.


예를 들어 저 숫자판에서 11과 33을 생각했다고 하자. 그렇다면 11과 33사이의 직각거리는 8이 된다.(동서방향거리 : 6, 남북방향거리 : 2)

- %4를 해서 나머지를 구해서 몇번째 줄에 있는지 확인해서 둘 사이의 세로 거리를 구한다.
- /4를해서 가로거리를 구한다.

```python
from IPython.display import Image
```

```python
def solution (A, B):
    A, B = A-1, B-1
    v = abs(A%4 - B%4)
    h = abs(A//4 - B//4)
    print(v, h)
    
```
